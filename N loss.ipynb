```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pickle

# 1. Load dataset
df = pd.read_excel('./Data/NF_runoff_input.xlsx')  # Replace with actual path

# 2. Rename columns
df = df.rename(columns={
    '年平均降雨量': 'rain',       # MAP
    '年平均温度': 'tem',         # MAT
    '土壤容重': 'bulk',         # BD
    '土壤粘粒含量%': 'clay',     # Clay
    '土壤有机碳SOC': 'soc',      # SOC
    '土壤全氮': 'tn',            # TN
    '土壤pH': 'pH',
    '坡度°': 'slope',
    '灌溉量mm': 'irr',          # IRR
    'NF_runoff': 'nf'           # target variable
})

# 3. Select features
features = ['tem', 'rain', 'bulk', 'clay', 'soc', 'tn', 'pH', 'slope', 'irr', 'nf']
df = df[features]

# 4. Clean data
df = df.apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)

# 5. Split data
X = df.drop(columns=['nf'])
y = df['nf']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6. Train model
param_grid = {'n_estimators': [250, 500, 800], 'max_depth': [10, 20, 30]}
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)

# 7. Evaluate
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Best Parameters:", grid_search.best_params_)
print("R²:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

# 8. (Optional) Save model
# with open('./Model/NF_runoff_model.pkl', 'wb') as f:
#     pickle.dump(best_model, f)
```
