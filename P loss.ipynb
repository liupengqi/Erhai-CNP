```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pickle

# 1. Load dataset
df = pd.read_excel('./Data/P_runoff.xlsx')  # Replace with your actual file path

# 2. Rename columns
df = df.rename(columns={
    '年平均降雨量': 'rain',
    '年平均温度': 'tem',
    '土壤容重': 'bulk',
    '土壤粘粒含量%': 'clay',
    '土壤有机碳SOC': 'soc',
    '土壤有效磷OP': 'op',
    '土壤总磷TP': 'tp',
    '土壤pH': 'pH',
    '坡度°': 'slope',
    '灌溉量mm': 'irr',
    '施肥是否优化': 'opt_fert',   # fertilizer optimization
    'Ploss': 'ploss'              # target
})

# 3. Encode fertilizer optimization: 是→1，否→0
df['opt_fert'] = df['opt_fert'].map({'是': 1, '否': 0}).fillna(df['opt_fert'])
df['opt_fert'] = pd.to_numeric(df['opt_fert'], errors='coerce')

# 4. Select features
features = ['rain', 'tem', 'bulk', 'clay', 'soc', 'op', 'tp', 'pH', 'slope', 'irr', 'opt_fert', 'ploss']
df = df[features]

# 5. Clean data
df = df.apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)

# 6. Train/test split
X = df.drop(columns=['ploss'])
y = df['ploss']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 7. Model training
param_grid = {'n_estimators': [250, 500, 800], 'max_depth': [10, 20, 30]}
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)

# 8. Evaluation
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Best Parameters:", grid_search.best_params_)
print("R²:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

# 9. (Optional) Save model
# with open('./Model/Ploss_RF_model.pkl', 'wb') as f:
#     pickle.dump(best_model, f)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pickle

# 1. Load dataset
df = pd.read_excel('./Data/PF_leaching.xlsx')  # Replace with your file path

# 2. Rename columns
df = df.rename(columns={
    '年平均降雨量': 'rain',
    '年平均温度': 'tem',
    '土壤容重': 'bulk',
    '土壤粘粒含量%': 'clay',
    '土壤有机碳SOC': 'soc',
    '土壤有效磷OP': 'op',
    '土壤总磷TP': 'tp',
    '土壤pH': 'pH',
    '坡度°': 'slope',
    '灌溉量mm': 'irr',
    'PF': 'pleaching'  # target
})

# 3. Select features
features = ['rain', 'tem', 'irr', 'bulk', 'clay', 'soc', 'op', 'tp', 'pH', 'slope', 'pleaching']
df = df[features]

# 4. Clean data
df = df.apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)

# 5. Train/test split
X = df.drop(columns=['pf'])
y = df['pf']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6. Model training
param_grid = {'n_estimators': [250, 500, 800], 'max_depth': [10, 20, 30]}
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)

# 7. Evaluation
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Best Parameters:", grid_search.best_params_)
print("R²:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

# 8. (Optional) Save model
# with open('./Model/PF_LF_model.pkl', 'wb') as f:
#     pickle.dump(best_model, f)
```
